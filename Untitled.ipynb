```python
import pandas as pd
import re
import matplotlib as plt
import numpy as np

sant = pd.read_csv("santander.csv") 
bank = pd.read_csv("bankinter.csv")
sab = pd.read_csv("banco-sabadell.csv")
bbva = pd.read_csv("bbva.csv")
caixa = pd.read_csv("caixabank.csv")
banks = [sant, bank, sab, bbva, caixa]
```

```python
def Date_digits(df):
    return int(df[:4])

for x in banks:
    x["Date"] = x["Date"].apply(Date_digits)
```

```python
def Cut_year(df):
    return df>2014

sant = sant[(sant["Date"].apply(Cut_year))]
bank = bank[(bank["Date"].apply(Cut_year))]
sab = sab[(sab["Date"].apply(Cut_year))]
bbva = bbva[(bbva["Date"].apply(Cut_year))]
caixa = caixa[(caixa["Date"].apply(Cut_year))]


sant["Bank"]="Santander"
bank["Bank"]="Bankinter"
sab["Bank"]="Banco Sabadell"
bbva["Bank"]="BBVA"
caixa["Bank"]="Caixa Bank"
```

```python
sant = sant.groupby(["Bank","Date"])["Close", "High", "Low", "Open", "Volume"].mean()
bank = bank.groupby(["Bank","Date"])["Close", "High", "Low", "Open", "Volume"].mean()
sab = sab.groupby(["Bank","Date"])["Close", "High", "Low", "Open", "Volume"].mean()
bbva = bbva.groupby(["Bank","Date"])["Close", "High", "Low", "Open", "Volume"].mean()
caixa = caixa.groupby(["Bank","Date"])["Close", "High", "Low", "Open", "Volume"].mean()
```

```python
import requests
from bs4 import BeautifulSoup

links=["https://www.marketwatch.com/investing/stock/san?countrycode=es",
      "https://www.marketwatch.com/investing/stock/bkt?countrycode=es",
      "https://www.marketwatch.com/investing/stock/sab?countrycode=es",
      "https://www.marketwatch.com/investing/stock/bbva?countrycode=es",
      "https://www.marketwatch.com/investing/stock/cabk?countrycode=es"]
names = ["Banco Santander","Banco Bankinter", "Banco Sabadell", "BBVA", "CaixaBank"]
```

```python
actual = []
previous = []
for i in range(len(links)):
    res = requests.get(links[i])
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    actual.append(soup("span", class_="value")[0].text)
    previous.append(soup("td",class_="table__cell u-semi")[0].text)
    print("\n{}: Actual stock closed value: {} €".format(names[i],actual[i]))
    print("{}: Previous stock closed value: {} €".format(names[i],previous[i]))
    print(soup("span",class_="timestamp__time")[0].text)
```
